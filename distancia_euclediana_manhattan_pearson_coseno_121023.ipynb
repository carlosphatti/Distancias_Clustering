### Distancia Euclediana

```python
x = [185, 170, 168, 179, 182, 188]
y = [72, 56, 60, 68, 72, 77]
```

```python
k1 = (185, 72)  # Punto de referencia clase 1
k2 = (170, 56)  # Punto de referencia clase 2
```

image.png

```python
def distEu(x1, y1, x2, y2):
    return ((x1 - x2)**2 + (y1 - y2)**2)**0.5#formula de distancia euclediana
```

```python
resultados = []
```

```python
for i in range(len(x)):
    dist_1 = distEu(x[i], y[i], k1[0], k1[1])#dist entre el punto actual y el punto de referencia k1
    dist_2 = distEu(x[i], y[i], k2[0], k2[1])##dist entre el punto actual y el punto de referencia k2


    clase = 1 if dist_1 < dist_2 else 2 #se asigna segun cual distancia sea menor
    resultados.append((x[i], y[i], dist_1, dist_2, clase))
```

### Distancia de Person

Ecuación_Correlación_de_Pearson.png

Calcular las medias de x y y

imagen.png

```python
x = [185, 170, 168, 179, 182, 188]
y = [72, 56, 60, 68, 72, 77]

# Calcular medias
media_x = sum(x) / len(x)
media_y = sum(y) / len(y)

print(f"Media de x: {media_x}")
print(f"Media de y: {media_y}")
```

La covarianza entre dos conjuntos de datos x e y

imagen.png

```python
# Calcular covarianza
cov_xy = sum((x[i] - media_x) * (y[i] - media_y) for i in range(len(x))) / len(x)

print(f"Covarianza entre x e y: {cov_xy}")
```

calcular la desviacion standar

imagen.png

```python
# Calcular desviaciones estándar
std_dev_x = (sum((xi - media_x)**2 for xi in x) / len(x))**0.5
std_dev_y = (sum((yi - media_y)**2 for yi in y) / len(y))**0.5

print(f"Desviación estándar de x: {std_dev_x}")
print(f"Desviación estándar de y: {std_dev_y}")
```

imagen.png

```python
# Calcular correlación de Pearson
correlacion = cov_xy / (std_dev_x * std_dev_y)

print(f"Correlación de Pearson (r): {correlacion}")
```

```python
# Graficar la matriz de correlación
plt.matshow(correlacion, cmap="bwr")
for i in range(len(correlacion)):
    for j in range(len(correlacion)):
        plt.text(i, j, round(correlacion[i, j], 2), ha="center", va="center")

plt.colorbar()
plt.title('Matriz de Correlación')
plt.xticks([0, 1], ['x', 'y'])
plt.yticks([0, 1], ['x', 'y'])
plt.show()
```

```python
# Función para calcular la distancia de coseno entre un punto y centroide k1 y k2
def distancia_coseno (puntos, centroide):
    dot_producto = sum(a*b for a, b in zip(puntos, centroide)) #se realiza el producto punto de la formula
    norm_punto = sum(a**2 for a in puntos) ** 0.5# normalizacion o magnitud del punto eu
    norm_centroide = sum(b**2 for b in centroide) ** 0.5
    # El producto escalar se divide por el producto de las normas de los vectores.
    return dot_producto / (norm_punto * norm_centroide)# formula de DISTANCIA DE COSENO

# Calcula las distancias de coseno y almacenar los resultados en una tabla
resultados = []

for i in range(len(x)):# se calculan las distancias de coseno a k1 y k2
    punto = ([x[i], y[i]])
    distancia_k1 = distancia_coseno (punto, k1)
    distancia_k2 = distancia_coseno (punto, k2)
    resultados.append([distancia_k1, distancia_k2])

# Función para calcular el ángulo en grados entre dos vectores
def angulo(puntos, centroide):
    coseno = distancia_coseno (puntos, centroide)
    angulo_rad = math.acos(coseno)
    return math.degrees(angulo_rad)

# Mostrar los resultados en una tabla con la columna del ángulo
print("Punto    | Distancia a k1 | Distancia a k2 | Ángulo (grados)")
print("--------------------------------------------------------")
for i in range(len(x)):
    angulo_k1 = angulo([x[i], y[i]], k1)
    angulo_k2 = angulo([x[i], y[i]], k2)
    print(f"Punto {i+1}  | {resultados[i][0]:.4f}          | {resultados[i][1]:.4f}          | {angulo_k1:.4f} / {angulo_k2:.4f}")
```
